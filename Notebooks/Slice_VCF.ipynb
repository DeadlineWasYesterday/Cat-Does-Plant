### `Slices original vcf file to keep only QTL markers`

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('WF1M20.vcf', sep = '\t', dtype = str)
```

```python
df2 = pd.read_csv('BLAST/340can range.csv')
```

```python
df3 = pd.DataFrame(columns = df.columns)

for q in list(df2['QTL'].unique()):
    tdf = df[df['#CHROM'] == df2['1'][df2['QTL'] == q].to_list()[0]].reset_index(drop = True)
    tdf['POS'] = tdf['POS'].astype(int)
    
    s,e = df2['Start'][df2['QTL'] == q].to_list()[0], df2['End'][df2['QTL'] == q].to_list()[0]
    
    print('QTL %s. %d to %d.' %(q,s,e))
    print('tdf shape %dx%d. df3 shape %dx%d.' % (tdf.shape + df3.shape))
    ti = []
    
    while s not in set(tdf['POS'].to_list()):
            s+=1
    while e not in set(tdf['POS'].to_list()):
        e-=1
    tdf3 = tdf.iloc[list(range(int(tdf[tdf['POS'] == s].index.values), int(tdf[tdf['POS'] == e].index.values + 1))), :]

    df3 = pd.concat([df3, tdf3], axis = 0).reset_index(drop = True)
```
