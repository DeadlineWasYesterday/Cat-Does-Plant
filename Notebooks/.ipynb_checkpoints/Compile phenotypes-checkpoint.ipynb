```python
df = pd.read_csv('../../Data/GWAS data/ori/Final_BD_screening_v2.csv')
dfc = df[df['Treatment'] == 'Control']
dfcs = dfc.groupby('accession_name').std()
dfc = dfc.groupby('accession_name').mean()
newcol = []
for col in dfcs.columns:
    newcol.append(col + ' std')
dfcs.columns = newcol
dfv = df.loc[:, ['accession_name', 'Total replicate', 'Variety', 'Variety.Name', 'Group']].reset_index(drop = True)
dfv = dfv.drop_duplicates('accession_name')
dfc = dfc.drop(columns = ['Replicate', 'Total replicate', 'Variety', 'SES'])
dfc.to_clipboard()
dfc = pd.read_clipboard()

dfcs = dfcs.drop(columns = ['Replicate std', 'Total replicate std', 'Variety std', 'SES std'])
dfc = pd.merge(dfc, dfv, on = 'accession_name', how = 'inner')
dfc = pd.merge(dfc, dfcs, on = 'accession_name', how = 'inner')
#dfc.to_csv('../Data/GWAS data/183x31 Control.csv', index = False)
```

```python
df = pd.read_csv('../Data/GWAS data/ori/Final_BD_screening_v2.csv')
dfs = df[df['Treatment'] == 'Stress']
dfss = dfs.groupby('accession_name').std()
dfs = dfs.groupby('accession_name').mean()
newcol = []
for col in dfss.columns:
    newcol.append(col + ' std')
dfss.columns = newcol
dfv = df[['accession_name', 'Total replicate', 'Variety', 'Variety.Name', 'Group']].copy()
dfv = dfv.drop_duplicates('accession_name')
dfs = dfs.drop(columns = ['Replicate', 'Total replicate', 'Variety'])
dfs.to_clipboard()
dfs = pd.read_clipboard()

dfss = dfss.drop(columns = ['Replicate std', 'Total replicate std', 'Variety std'])
dfs = pd.merge(dfs, dfv, on = 'accession_name', how= 'left')
dfs = pd.merge(dfs, dfss, on = 'accession_name', how = 'left')
#dfs.to_csv('../Data/GWAS data/183x33 Stress.csv', index = False)
```

```python
df2['prwl'] = (df['Root weight'] - df['Stress Root weight']) / df['Root weight']
df2['pswl'] = (df['Shoot weight'] - df['Stress Shoot weight']) / df['Shoot weight']
df2['prll'] = (df['Root length'] - df['Stress Root length']) / df['Root length']
df2['psll'] = (df['Shoot length'] - df['Stress Shoot length']) / df['Shoot length']
df2['prng'] = (df['Stress Root Na+'] - df['Root Na+']) / df['Root Na+']
df2['psng'] = (df['Stress Shoot Na+'] - df['Shoot Na+']) / df['Shoot Na+']
df2['pcal'] = (df['Chlorophyll A'] - df['Stress Chlorophyll A']) / df['Chlorophyll A']
df2['pcbl'] = (df['Chlorophyll B'] - df['Stress Chlorophyll B']) / df['Chlorophyll B']
df2['SES'] = df['SES']
```
