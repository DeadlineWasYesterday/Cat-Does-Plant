```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import seaborn as sns
sns.set()
%matplotlib notebook
```

```python
pd.read_excel('S3 Table Observed mean phenotype values.xlsx', header = 1).columns
```

```python
t = pd.read_excel('S3 Table Observed mean phenotype values.xlsx', header = 1)[['accession_name', 'Shoot_weight', 'Chlorophyll_A',
                                                                          'Stress_Shoot_weight', 'Stress_Chlorophyll_A']]
```

```python
t.iloc[:,[1,3]].plot(kind = 'kde')
```

```python
t.iloc[:,[2,4]].plot(kind = 'kde')
```

```python
t.iloc[:,[1,2]].plot(kind = 'scatter', x = 'Shoot_weight', y = 'Chlorophyll_A')
z = np.polyfit(t['Shoot_weight'].to_list(), t['Chlorophyll_A'].to_list(), 1)
p = np.poly1d(z)
plt.plot(t['Shoot_weight'].to_list(),p(t['Shoot_weight'].to_list()),"r--")

plt.show()
```

```python
t.iloc[:,[3,4]].plot(kind = 'scatter', x = 'Stress_Shoot_weight', y = 'Stress_Chlorophyll_A')
z = np.polyfit(t['Stress_Shoot_weight'].to_list(), t['Stress_Chlorophyll_A'].to_list(), 1)
p = np.poly1d(z)
plt.plot(t['Stress_Shoot_weight'].to_list(),p(t['Stress_Shoot_weight'].to_list()),"r--")

plt.show()
```

```python
t['RC'] = t['Shoot_weight'] / t['Chlorophyll_A']
t['RS'] = t['Stress_Shoot_weight'] / t['Stress_Chlorophyll_A']
```
